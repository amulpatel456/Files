# Classification and Regression technique to dataset for predict target variable

# 1.Linear Regression

```python
#import the libraries 

import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 

%matplotlib inline
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#check if any null value
data.isnull().sum()
```

```python
#correlation matrix 

correlation_matrix = data.corr().round(2)
print(correlation_matrix)
```

```python
feature = data['TotalUsers']
target = data['MaleCyc_Total']
plt.scatter(feature,target, marker='o')
plt.title('TotalUsers and MaleCyc_Total')
plt.xlabel('Total User')
plt.ylabel('MaleCyc_Total')
```

# Model Building

```python
#data for preparation

x=data[['TotalUsers']]
y=data['MaleCyc_Total']
```

```python
# splits the training and test data set in 80% : 20%
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# linear regression model

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

# Prediction

```python
#predicted variables and plot the regression line

Y_pred = linear_model.predict(X_test)
plt.scatter(X_test, Y_test)
plt.plot(X_test, Y_pred, color ='r')
plt.title("TotalUsers and MaleCyc_Total")
plt.xlabel("Independent variable")
plt.ylabel("Dependent variable")
```

# Performance

```python
#performance of the linear model

print(linear_model.score(X_test, Y_test))
```

```python
print(mean_squared_error(Y_test,Y_pred))
```

# 2. Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn import metrics
```

```python
#correlation matrix

correlation_matrix = data.corr().round(2)
print(correlation_matrix)
```

```python
# independent variable and the target y value

x= data[['TotalUsers']]
y = data.MaleCyc_Total

```

```python
# min and max

print(x.max())
print(x.min())
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X = pd.DataFrame(x_scaled)
```

```python
#80% is train data and 20% is test data

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train,y_train)
```

```python
# prediction for the test data

y_pred=logistic_regression.predict(X_test)
```

```python
#visualising the graph

for i in range(1):
    plt.figure(i, figsize=(4, 3))
    plt.title("logistic regression")
    plt.xlabel('TotalUsers')
    ax=sns.regplot(x=X_test[i], y=y_pred, data=data, logistic=True,ci=None,line_kws={'color':'red'})
    ax.set(xlabel='TotalUsers',ylabel='MaleCyc_Total')
    plt.show()
```

```python
#confusion matrix

confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
#visualising the confusion matrix by creatig a heatmap
sns.heatmap(pd.DataFrame(confusion_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

# Performance

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
